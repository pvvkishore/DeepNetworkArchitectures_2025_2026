### Dataset (Learning Data)
We will teach a machine to learn a mapping from:

\[
X=[-1,0,1,2,3],\quad y=[1,2,3,4,5]
\]

We'll assume the model family:

\[
\hat{y} = wx + b
\]

and learn \(w\) and \(b\) by minimizing the Mean Squared Error with gradient descent.


<div style="background:#fef3c7; padding:14px; border-radius:12px; border: 1px solid #f59e0b;">
  <b>Learning objective (Cost Function)</b><br>
  We measure "how wrong" the model is using:
</div>


```python
def predict(X, w, b):
    return w * X + b
# ERROR or LOSS - > COST FUnction - It is Mean square error
def cost(X, y, w, b):
    y_hat = predict(X, w, b)
    return (1/(2*len(X))) * ((y_hat - y)**2).sum()

print("Initial cost J(0,0) =", cost(X, y, 0.0, 0.0))
```


<div style="background:#dcfce7; padding:14px; border-radius:12px; border: 1px solid #22c55e;">
  <b>Gradients - Optimization - Stochastic Gradient Descent Algorithm</b><br>
  The direction to adjust parameters is given by partial derivatives:
</div>

```python
def gradients(X, y, w, b):
    y_hat = predict(X, w, b)
    error = y_hat - y
    dw = (1/len(X)) * np.sum(error * X)
    db = (1/len(X)) * np.sum(error)
    return dw, db

dw0, db0 = gradients(X, y, 0.0, 0.0)
print("Gradients at (w=0, b=0):")
print("  dJ/dw =", dw0)
print("  dJ/db =", db0)
```

### Visualizing the starting model

At \(w=0, b=0\), the model predicts a flat line at 0.

```python
w, b = 0.0, 0.0

plt.figure()
plt.scatter(X, y, label="Data")
x_line = np.linspace(X.min()-0.5, X.max()+0.5, 200)
plt.plot(x_line, predict(x_line, w, b), label="Initial model")
plt.title("Initial predictions (w=0, b=0)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.25)
plt.show()
```


<div style="background:#e0f2fe; padding:14px; border-radius:12px; border: 1px solid #38bdf8;">
  <b>Gradient Descent Loop</b><br>
  We'll apply the update rule repeatedly and watch the model improve.
</div>


```python
def gradient_descent(X, y, w0=0.0, b0=0.0, alpha=0.2, n_iters=20):
    w, b = w0, b0
    history = []

    for i in range(n_iters + 1):
        J = cost(X, y, w, b)
        y_hat = predict(X, w, b)
        dw, db = gradients(X, y, w, b)

        history.append({
            "iter": i,
            "w": w,
            "b": b,
            "cost": J,
            "dw": dw,
            "db": db,
            "y_hat": y_hat.copy()
        })

        if i < n_iters:
            w = w - alpha * dw
            b = b - alpha * db

    return history

history = gradient_descent(X, y, alpha=0.2, n_iters=20)

# Show early iterations
pd.DataFrame([{k: v for k, v in h.items() if k in ("iter", "w", "b", "cost", "dw", "db")} for h in history]).head(6)
```

### Model snapshots across iterations

We'll plot the fitted line at a few iterations to see the learning movement.

```python
def plot_model_at(iter_idx):
    h = history[iter_idx]
    w, b = h["w"], h["b"]

    plt.figure()
    plt.scatter(X, y, label="Data")
    x_line = np.linspace(X.min()-0.5, X.max()+0.5, 200)
    plt.plot(x_line, predict(x_line, w, b), label=f"Model @ iter {iter_idx}")
    plt.title(f"Model at iteration {iter_idx}  (w={w:.3f}, b={b:.3f})")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True, alpha=0.25)
    plt.show()

for idx in [0, 1, 2, 3, 5, 10, 20]:
    plot_model_at(idx)
```

### Cost falling over time

A key sign of learning is that the cost should decrease.

```python
iters = [h["iter"] for h in history]
costs = [h["cost"] for h in history]

plt.figure()
plt.plot(iters, costs, marker="o")
plt.title("Cost J(w,b) vs Iteration")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.grid(True, alpha=0.25)
plt.show()
```

<div style="background:#f3e8ff; padding:14px; border-radius:12px; border: 1px solid #a855f7;">
  <b>Final learned parameters (from gradient descent)</b>
</div>

```python
final = history[-1]
w_final, b_final = final["w"], final["b"]

print("Final parameters:")
print("  w =", w_final)
print("  b =", b_final)
print("Final cost:", final["cost"])

y_pred = predict(X, w_final, b_final)

result_df = pd.DataFrame({
    "x": X,
    "y_true": y,
    "y_pred": np.round(y_pred, 4),
    "error": np.round(y_pred - y, 4)
})
result_df
```

```python
plt.figure()
plt.scatter(X, y, label="Data")
x_line = np.linspace(X.min()-0.5, X.max()+0.5, 200)
plt.plot(x_line, predict(x_line, w_final, b_final), label="Learned model")
plt.title("Final model after gradient descent")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.25)
plt.show()
```


### Bonus: Try different learning rates

Learning rate \( \alpha \) controls step size:
- Too small â†’ slow learning  
- Too large â†’ may overshoot or diverge  

Experiment by changing \( \alpha \) and re-running the cell below.


```python

def run_with_alpha(alpha):
    h = gradient_descent(X, y, alpha=alpha, n_iters=20)
    return h

for alpha in [0.05, 0.2, 0.6]:
    h = run_with_alpha(alpha)
    print(f"alpha={alpha:<4} -> w={h[-1]['w']:.4f}, b={h[-1]['b']:.4f}, cost={h[-1]['cost']:.6f}")

plt.figure()
for alpha in [0.05, 0.2, 0.6]:
    h = run_with_alpha(alpha)
    plt.plot([x["iter"] for x in h], [x["cost"] for x in h], marker="o", label=f"alpha={alpha}")

plt.title("Cost curves for different learning rates")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.legend()
plt.grid(True, alpha=0.25)
plt.show()
```
